```python
import numpy as np
import matplotlib.pyplot as plt

# Data
data_di = [
    [1.0, 1.1, 1.25],
    [1.0, 1.3, 1.8],
    [1.0, 1.5, 2.5]
]

x = np.array([1, 2, 3])
x_dense = np.linspace(1, 3, 200)

# Plot
plt.figure(figsize=(8, 5))

for idx, y_vals in enumerate(data_di):
    y = np.array(y_vals)
    coeffs = np.polyfit(x, y, 2)
    poly = np.poly1d(coeffs)
    y_dense = poly(x_dense)

    plt.plot(x_dense, y_dense, label=f'Scenario {idx + 1}')
    plt.scatter(x, y, color=plt.gca().lines[-1].get_color(), edgecolors='k')

# Customizations
plt.xlabel('m')
plt.ylabel('$DI_m$')
plt.title('Demand increase due to transportation method selection')
plt.xticks([1, 2, 3])
plt.legend()
plt.grid(alpha=0.3)
plt.tight_layout()

# ✅ Save figure BEFORE showing it
plt.savefig('demand_increase_plot.png', dpi=300)
plt.show()
```

```python
import numpy as np
import pandas as pd
import random

# --- D_kij 계산 시작 ---
products = ['k1','k2','k3']
plant_idx = range(3)
plant_index = pd.MultiIndex.from_product([products, plant_idx], names=['k','i'])
dc_labels = dc_candidates.index.tolist()

D_ij = pd.DataFrame(index=plant_index, columns=dc_labels, dtype=float)
for k, i in plant_index:
    px, py = plant_candidates.loc[k, i]
    for j in dc_labels:
        dx, dy = dc_candidates.loc[j, 'point']
        D_ij.loc[(k, i), j] = np.hypot(px - dx, py - dy)

flat_df = (
    D_ij
    .reset_index()            # ('k','i')를 컬럼으로 풀어내고
    .melt(id_vars=['k','i'], var_name='j', value_name='D_kij')
)

# --- j를 숫자 기준으로 정렬하기 위해 숫자 칼럼(j_num) 추가 ---
flat_df['j_num'] = flat_df['j'].str.extract(r'(\d+)').astype(int)

# k, i 순서를 그대로 두되, j_numeric 순으로 오름차순 정렬
flat_df = flat_df.sort_values(by=['k','i','j_num'])

# 최종 열 순서를 k, i, j, D_kij 순으로 재배치하고, 정렬을 위해 사용했던 j_num은 삭제
flat_df = flat_df[['k','i','j','D_kij']]

# 엑셀로 저장
flat_df.to_excel('2.xlsx', index=False)

D_ij
```

```python
import numpy as np
import pandas as pd
import random
import os

# --- 1. 데이터 생성: s_rk (시트 '17' 용) ---
np.random.seed(252432342)
matrix = np.random.randint(0, 2, size=(50, 3))
# 모든 행에 최소 하나 이상의 '1'이 있도록 보장
for i in range(matrix.shape[0]):
    if np.all(matrix[i] == 0):
        random_index = np.random.randint(0, 3)
        matrix[i, random_index] = 1
row_labels_s_rk = [f"r{i+1}" for i in range(50)]
col_labels_s_rk = [f"k{j+1}" for j in range(3)]
s_rk = pd.DataFrame(matrix, index=row_labels_s_rk, columns=col_labels_s_rk)


# --- 2. 데이터 생성: plant_candidates (시트 'plant' 용) ---
random.seed(24535)
all_points = [
    (x, y)
    for x in range(0, 1001)
    for y in range(0, 1001)
    if not (200 <= x < 800 and 200 <= y < 800)
]
selected_points = random.sample(all_points, 9)
random.shuffle(selected_points)
plant_candidates = pd.DataFrame(
    [selected_points[:3], selected_points[3:6], selected_points[6:]],
    index=['k1', 'k2', 'k3']
)


# --- 3. 데이터 생성: customer_location (시트 'customer' 용) ---
np.random.seed(2255)
x_coords = np.random.normal(loc=500, scale=200, size=50)
y_coords = np.random.normal(loc=500, scale=200, size=50)
x_int = np.clip(np.round(x_coords), 0, 1000).astype(int)
y_int = np.clip(np.round(y_coords), 0, 1000).astype(int)
points = list(zip(x_int, y_int))
row_labels_customer = [f"r{i+1}" for i in range(50)]
customer_location = pd.DataFrame({'point': points}, index=row_labels_customer)

# --- 4. 데이터 생성: dc_candidates (시트 'dc' 용) ---
random.seed(2594532456)

# "도넛 모양" 외부 영역 정의
all_points_dc = [
    (x, y)
    for x in range(0, 1001)
    for y in range(0, 1001)
    if not (200 <= x < 800 and 200 <= y < 800)
]

# 10개 무작위 샘플링
selected_points_dc = random.sample(all_points_dc, 10)

# 행 레이블 생성
row_labels_dc = [f"dc{i+1}" for i in range(10)]

# DataFrame에 저장
dc_candidates = pd.DataFrame({'point': selected_points_dc}, index=row_labels_dc)

# --- 5. 데이터 생성: mu_df (시트 '6' 용) ---
# Set seed
np.random.seed(439343)

# 50x3 크기의 데이터 생성 (수정됨)
mu_matrix = 10 * np.random.uniform(1, 5, size=(50, 3))

# DataFrame으로 변환
mu_df = pd.DataFrame(mu_matrix, columns=['product1', 'product2', 'product3'])

# --- 6. 데이터 생성: mu_hat_df (시트 '5' 용) ---
# Set seed
np.random.seed(74534)

U_matrix = np.random.uniform(4, 10, size=(50, 3))

# mu_k와 U[4,10]의 원소별 최소값을 취하여 mu_hat_matrix 생성
mu_hat_matrix = np.minimum(mu_matrix, U_matrix)

# DataFrame으로 변환
mu_hat_df = pd.DataFrame(mu_hat_matrix, columns=['product1', 'product2', 'product3'])

# --- 7. 데이터 생성: flat_df (시트 '2' 용) ---
products = ['k1','k2','k3']
plant_idx = range(3)
plant_index = pd.MultiIndex.from_product([products, plant_idx], names=['k','i'])
dc_labels = dc_candidates.index.tolist()

D_ij = pd.DataFrame(index=plant_index, columns=dc_labels, dtype=float)
for k, i in plant_index:
    px, py = plant_candidates.loc[k, i]
    for j in dc_labels:
        dx, dy = dc_candidates.loc[j, 'point']
        D_ij.loc[(k, i), j] = np.hypot(px - dx, py - dy)

flat_df = (
    D_ij
    .reset_index()
    .melt(id_vars=['k','i'], var_name='j', value_name='D_kij')
)

# j를 숫자 기준으로 정렬하기 위해 숫자 칼럼(j_num) 추가
flat_df['j_num'] = flat_df['j'].str.extract(r'(\d+)').astype(int)

# k, i 순서를 그대로 두되, j_numeric 순으로 오름차순 정렬
flat_df = flat_df.sort_values(by=['k','i','j_num'])

# 최종 열 순서를 k, i, j, D_kij 순으로 재배치하고, 정렬을 위해 사용했던 j_num은 삭제
flat_df = flat_df[['k','i','j','D_kij']]

# --- 8. 데이터 생성: D_jr (시트 '3' 용) ---
# 1) Extract coordinates into numpy arrays
cust_points = np.vstack(customer_location['point'].values)
dc_points   = np.vstack(dc_candidates['point'].values)

# 2) Compute distance matrix via broadcasting
diffs = dc_points[:, None, :] - cust_points[None, :, :]
dist_mat = np.sqrt(np.sum(diffs**2, axis=2))

# 3) Wrap into a DataFrame
D_jr = pd.DataFrame(
    dist_mat,
    index=dc_candidates.index,
    columns=customer_location.index
)

# --- 9. 데이터 생성: sheet_7_df (시트 '7' 용) ---
# Set seed for reproducibility
np.random.seed(34567) # 새로운 난수 시드 설정

# 엑셀 공식 =(30.74193*50*(1 + (RAND() - 0.5)))를 파이썬으로 구현
# 3x3 크기의 난수 배열 생성
random_values = 30.74193 * 50 * (1 + (np.random.rand(3, 3) - 0.5))

# DataFrame으로 변환
sheet_7_df = pd.DataFrame(random_values)

# --- 10. 데이터 생성: sheet_8_df (시트 '8' 용) ---
# Set seed for reproducibility
np.random.seed(88888) # 새로운 난수 시드 설정

# 엑셀 공식 =50*30.21582*(1 + (RAND() - 0.5))를 파이썬으로 구현
# 10x1 크기의 난수 배열 생성
random_values_8 = 50 * 30.21582 * (1 + (np.random.rand(10, 1) - 0.5))

# DataFrame으로 변환
sheet_8_df = pd.DataFrame(random_values_8)

# --- 11. 데이터 생성: sheet_9_df (시트 '9' 용) ---
# Set seed for reproducibility
np.random.seed(99999) # 새로운 난수 시드 설정

# 엑셀 공식 =10000*29.63441*(1 + (RAND() - 0.5))를 파이썬으로 구현
# 3x3 크기의 난수 배열 생성
random_values_9 = 10000 * 29.63441 * (1 + (np.random.rand(3, 3) - 0.5))

# DataFrame으로 변환
sheet_9_df = pd.DataFrame(random_values_9)

# --- 12. 데이터 생성: sheet_10_df (시트 '10' 용) ---
# Set seed for reproducibility
np.random.seed(101010) # 새로운 난수 시드 설정

# 엑셀 공식 =1000*30.21582*(1 + (RAND() - 0.5))를 파이썬으로 구현
# 10x1 크기의 난수 배열 생성
random_values_10 = 1000 * 30.21582 * (1 + (np.random.rand(10, 1) - 0.5))

# DataFrame으로 변환
sheet_10_df = pd.DataFrame(random_values_10)

# --- 13. 데이터 생성: sheet_11_df (시트 '11' 용) ---
# Set seed for reproducibility
np.random.seed(111111) # 새로운 난수 시드 설정

# flat_df의 D_kij 값(90개)에 랜덤 요소를 적용
# len(flat_df)를 사용하여 D_kij 값의 개수(90)만큼 난수 생성
random_d_kij = flat_df['D_kij'] * (1 + (np.random.rand(len(flat_df)) - 0.5)) * 0.5

# DataFrame으로 변환
sheet_11_df = pd.DataFrame(random_d_kij)

# --- 14. 데이터 생성: sheet_12_df (시트 '12' 용) ---
# Set seed for reproducibility
np.random.seed(121212) # 새로운 난수 시드 설정

# D_jr 값(10x50)에 랜덤 요소를 적용
# D_jr.shape를 사용하여 D_jr의 크기와 동일한 난수 배열 생성
sheet_12_df = D_jr * (1 + (np.random.rand(*D_jr.shape) - 0.5)) * 0.5

# --- 15. 데이터 생성: sheet_13_df (시트 '13' 용) ---
# Set seed for reproducibility
np.random.seed(131313) # 새로운 난수 시드 설정

# 엑셀 공식 =100 * 30.21582 * (1 + (RAND() - 0.5))를 파이썬으로 구현
# 10x1 크기의 난수 배열 생성
random_values_13 = 100 * 30.21582 * (1 + (np.random.rand(10, 1) - 0.5))

# DataFrame으로 변환
sheet_13_df = pd.DataFrame(random_values_13)

# --- 16. 데이터 생성: sheet_14_df (시트 '14' 용) ---
# Set seed for reproducibility
np.random.seed(141414) # 새로운 난수 시드 설정

# 엑셀 공식 =10 * (1 + (RAND() - 0.5))를 파이썬으로 구현
# 10x1 크기의 난수 배열 생성
random_values_14 = 10 * (1 + (np.random.rand(10, 1) - 0.5))

# DataFrame으로 변환
sheet_14_df = pd.DataFrame(random_values_14)

# --- 17. 데이터 생성: sheet_15_df (시트 '15' 용) ---
# Set seed for reproducibility
np.random.seed(151515) # 새로운 난수 시드 설정

# 엑셀 공식 =(1+ RAND() * 0.4 - 0.2) * 500를 파이썬으로 구현
# 3x3 크기의 난수 배열 생성
random_values_15 = (1 + np.random.rand(3, 3) * 0.4 - 0.2) * 500

# DataFrame으로 변환
sheet_15_df = pd.DataFrame(random_values_15)

################################################################
################################################################
################################################################
# --- 4. 엑셀 파일 한 번에 저장하기 ---
file_name = 'data.xlsx'

print(f"'{file_name}' 파일 생성을 시작합니다...")

with pd.ExcelWriter(file_name, engine='openpyxl') as writer:
    # 시트 '17'에 s_rk 데이터 저장 (A2:C51)
    s_rk.to_excel(
        writer,
        sheet_name='17',
        startrow=1,
        startcol=0,
        header=False,
        index=False
    )
    print(" -> '17' 시트 저장 완료")

    # 시트 'plant'에 plant_candidates 데이터 저장 (A1:C3)
    plant_candidates.to_excel(
        writer,
        sheet_name='plant',
        startrow=0,
        startcol=0,
        header=False,
        index=False
    )
    print(" -> 'plant' 시트 저장 완료")

    # 시트 'customer'에 customer_location 데이터 저장 (A1:A50)
    customer_location.to_excel(
        writer,
        sheet_name='customer',
        startrow=0,
        startcol=0,
        header=False,
        index=False
    )
    print(" -> 'customer' 시트 저장 완료")

    # 시트 'dc'에 dc_candidates 데이터 저장 (A1:A10)
    dc_candidates.to_excel(
        writer,
        sheet_name='dc',
        startrow=0,
        startcol=0,
        header=False,
        index=False
    )
    print(" -> 'dc' 시트 저장 완료")

    # 시트 '6'에 mu_df 데이터 저장 (A2:C51)
    mu_df.to_excel(
        writer,
        sheet_name='6',
        startrow=1,
        startcol=0,
        header=False,
        index=False
    )
    print(" -> '6' 시트 저장 완료")

    # 시트 '5'에 mu_hat_df 데이터 저장 (A2:C51)
    mu_hat_df.to_excel(
        writer,
        sheet_name='5',
        startrow=1,
        startcol=0,
        header=False,
        index=False
    )
    print(" -> '5' 시트 저장 완료")

    # 시트 '2'에 flat_df 데이터 저장 (A1:D91)
    flat_df.to_excel(
        writer,
        sheet_name='2',
        startrow=0,
        startcol=0,
        header=True,   # 열 이름(k,i,j,D_kij) 포함
        index=False
    )
    print(" -> '2' 시트 저장 완료")

    # 시트 '3'에 D_jr 데이터 저장 (A1:AX11)
    D_jr.to_excel(
        writer,
        sheet_name='3',
        startrow=0,
        startcol=0,
        header=True,  # 열 이름(r1..r50) 포함
        index=False   # 행 이름(dc1..dc10) 제외
    )
    print(" -> '3' 시트 저장 완료")

    # 시트 '7'에 sheet_7_df 데이터 저장 (B2:D4)
    sheet_7_df.to_excel(
        writer,
        sheet_name='7',
        startrow=1,   # 엑셀 2행
        startcol=1,   # 엑셀 B열
        header=False,
        index=False
    )
    print(" -> '7' 시트 저장 완료")

    # 시트 '8'에 sheet_8_df 데이터 저장 (B1:B10)
    sheet_8_df.to_excel(
        writer,
        sheet_name='8',
        startrow=0,   # 엑셀 1행
        startcol=1,   # 엑셀 B열
        header=False,
        index=False
    )
    print(" -> '8' 시트 저장 완료")

    # 시트 '9'에 sheet_9_df 데이터 저장 (B2:D4)
    sheet_9_df.to_excel(
        writer,
        sheet_name='9',
        startrow=1,   # 엑셀 2행
        startcol=1,   # 엑셀 B열
        header=False,
        index=False
    )
    print(" -> '9' 시트 저장 완료")

    # 시트 '10'에 sheet_10_df 데이터 저장 (B1:B10)
    sheet_10_df.to_excel(
        writer,
        sheet_name='10',
        startrow=0,   # 엑셀 1행
        startcol=1,   # 엑셀 B열
        header=False,
        index=False
    )
    print(" -> '10' 시트 저장 완료")

    # 시트 '11'에 random_d_kij 데이터 저장 (D2:D91)
    sheet_11_df.to_excel(
        writer,
        sheet_name='11',
        startrow=1,   # 엑셀 2행
        startcol=3,   # 엑셀 D열
        header=False,
        index=False
    )
    print(" -> '11' 시트 저장 완료")

    # 시트 '12'에 sheet_12_df 데이터 저장 (A2:AX11)
    sheet_12_df.to_excel(
        writer,
        sheet_name='12',
        startrow=1,   # 엑셀 2행
        startcol=0,   # 엑셀 A열
        header=False,
        index=False
    )
    print(" -> '12' 시트 저장 완료")

    # 시트 '13'에 sheet_13_df 데이터 저장 (B2:B11)
    sheet_13_df.to_excel(
        writer,
        sheet_name='13',
        startrow=1,   # 엑셀 2행
        startcol=1,   # 엑셀 B열
        header=False,
        index=False
    )
    print(" -> '13' 시트 저장 완료")

    # 시트 '14'에 sheet_14_df 데이터 저장 (B2:B11)
    sheet_14_df.to_excel(
        writer,
        sheet_name='14',
        startrow=1,   # 엑셀 2행
        startcol=1,   # 엑셀 B열
        header=False,
        index=False
    )
    print(" -> '14' 시트 저장 완료")

    # 시트 '15'에 sheet_15_df 데이터 저장 (B2:D4)
    sheet_15_df.to_excel(
        writer,
        sheet_name='15',
        startrow=1,   # 엑셀 2행
        startcol=1,   # 엑셀 B열
        header=False,
        index=False
    )
    print(" -> '15' 시트 저장 완료")

print(f"✅ '{file_name}' 파일에 모든 시트 저장을 완료했습니다.")
```

```python
import numpy as np
import pandas as pd
import random
import os
from scipy.stats import truncnorm

P = 5000

for i in range(1, 31):
  file_name = f'/content/drive/MyDrive/First Paper/newdata/data_{i}.xlsx'

  # s_rk (sheet '17') ---
  matrix = np.random.randint(0, 2, size=(100, 3))
  for i in range(matrix.shape[0]):
      if np.all(matrix[i] == 0):
          matrix[i, np.random.randint(0, 3)] = 1
  s_rk = pd.DataFrame(matrix, index=[f"r{i+1}" for i in range(100)], columns=[f"k{j+1}" for j in range(3)])

  # plant_candidates (sheet 'plant') ---
  all_points = [(x, y) for x in range(1001) for y in range(1001) if not (200 <= x < 800 and 200 <= y < 800)]
  selected_points = random.sample(all_points, 9)
  random.shuffle(selected_points)
  plant_candidates = pd.DataFrame([selected_points[:3], selected_points[3:6], selected_points[6:]], index=['k1', 'k2', 'k3'])

  # customer_location (sheet 'customer') ---
  x = np.clip(np.round(np.random.normal(500, 200, 100)), 0, 1000).astype(int)
  y = np.clip(np.round(np.random.normal(500, 200, 100)), 0, 1000).astype(int)
  customer_location = pd.DataFrame({'point': list(zip(x, y))}, index=[f"r{i+1}" for i in range(100)])

  # dc_candidates (sheet 'dc') ---
  all_points_dc = [(x, y) for x in range(1001) for y in range(1001) if not (200 <= x < 800 and 200 <= y < 800)]
  dc_candidates = pd.DataFrame({'point': random.sample(all_points_dc, 10)}, index=[f"dc{i+1}" for i in range(10)])

  # mu_df (sheet '6') ---
  mu_matrix = 10 * np.random.uniform(1, 5, size=(100, 3))
  mu_df = pd.DataFrame(mu_matrix, columns=['product1', 'product2', 'product3'])

  D = mu_df.values.mean()

  column_means = mu_df.mean()
  D1 = column_means["product1"]
  D2 = column_means["product2"]
  D3 = column_means["product3"]

  # mu_hat_df (sheet '5') ---
  U_matrix = np.random.uniform(4, 10, size=(100, 3))
  mu_hat_df = pd.DataFrame(np.minimum(mu_matrix, U_matrix), columns=['product1', 'product2', 'product3'])

  # flat_df (sheet '2') --- ij distance
  plant_index = pd.MultiIndex.from_product([['k1','k2','k3'], range(3)], names=['k','i'])
  D_ij = pd.DataFrame(index=plant_index, columns=dc_candidates.index, dtype=float)
  for k, i in plant_index:
      px, py = plant_candidates.loc[k, i]
      for j in dc_candidates.index:
          dx, dy = dc_candidates.loc[j, 'point']
          D_ij.loc[(k, i), j] = np.hypot(px - dx, py - dy)
  flat_df = D_ij.reset_index().melt(id_vars=['k','i'], var_name='j', value_name='D_kij')
  flat_df['j_num'] = flat_df['j'].str.extract(r'(\d+)').astype(int)
  flat_df = flat_df.sort_values(by=['k','i','j_num'])[['k','i','j','D_kij']]

  # D_jr (sheet '3') --- jr distance
  cust_points = np.vstack(customer_location['point'].values)
  dc_points = np.vstack(dc_candidates['point'].values)
  dist_mat = np.sqrt(np.sum((dc_points[:, None, :] - cust_points[None, :, :]) ** 2, axis=2))
  D_jr = pd.DataFrame(dist_mat, index=dc_candidates.index, columns=customer_location.index)

  # sheet_7_df (sheet '7') --- #MP
  d_means = np.array([D1, D2, D3]).reshape(3,1)
  sheet_7_df = pd.DataFrame(60 * d_means * (1 + (np.random.rand(3, 3) - 0.5)))

  # sheet_8_df (sheet '8') --- #MC
  sheet_8_df = pd.DataFrame(90 * D * (1 + (np.random.rand(10, 1) - 0.5)))

  # sheet_9_df (sheet '9') --- #factory construction
  sheet_9_df = pd.DataFrame((1/10) * P * sheet_7_df)

  # sheet_10_df (sheet '10') --- # DC construction
  sheet_10_df = pd.DataFrame((1/20) * P * sheet_8_df)

  # sheet_11_df (sheet '11') --- #ij route fixed
  sheet_11_df = pd.DataFrame(flat_df['D_kij'] * (1 + (np.random.rand(len(flat_df)) - 0.5)) * 100)

  # sheet_12_df (sheet '12') --- #jr route fixed
  sheet_12_df = D_jr * (1 + (np.random.rand(*D_jr.shape) - 0.5)) * 100

  # sheet_13_df (sheet '13') --- #f_j
  sheet_13_df = pd.DataFrame(D * P * (1 + (np.random.rand(10, 1) - 0.5)))

  # sheet_14_df (sheet '14') --- #h_j
  sheet_14_df = pd.DataFrame(0.2 * P * (1 + (np.random.rand(10, 1) - 0.5)))

  # sheet_15_df (sheet '15') --- #production cost
  sheet_15_df = pd.DataFrame((1 + np.random.rand(3, 3) * 0.4 - 0.2) * 1/3 * P)

  # sheet_demand_df (sheet 'demand') ---
  demand_rows_LD = []
  for _ in range(3):
      k = np.random.uniform(0, np.log(3) / 6)  #0~ln3/2 사이에서 3가지 케이스, 낮을 때, 높을 때, 섞여있을 때 => 높을 때랑 섞여있을 때 우선 해보면 될듯
      b = np.exp(k)
      c = np.exp(2 * k)
      demand_rows_LD.append([1.0, round(b, 2), round(c, 2)])

  # DataFrame으로 변환
  sheet_demand_LD_df = pd.DataFrame(demand_rows_LD)

  demand_rows_MD = []

  # k를 추출할 세 가지 범위 리스트
  intervals = [
      (0, np.log(3) / 6),
      (np.log(3) / 6, np.log(3) / 3),
      (np.log(3) / 3, np.log(3) / 2)
  ]

  # 각 범위(interval)에서 k를 하나씩 추출
  for low, high in intervals:
      k = np.random.uniform(low, high)
      b = np.exp(k)
      c = np.exp(2 * k)
      demand_rows_MD.append([1.0, round(b, 2), round(c, 2)])

  # DataFrame으로 변환
  sheet_demand_MD_df = pd.DataFrame(demand_rows_MD)

  # sheet_demand_df (sheet 'demand') ---
  demand_rows_HD = []
  for _ in range(3):
      k = np.random.uniform(np.log(3)/3, np.log(3) / 2)  #0~ln3/2 사이에서 3가지 케이스, 낮을 때, 높을 때, 섞여있을 때 => 높을 때랑 섞여있을 때 우선 해보면 될듯
      b = np.exp(k)
      c = np.exp(2 * k)
      demand_rows_HD.append([1.0, round(b, 2), round(c, 2)])

  # DataFrame으로 변환
  sheet_demand_HD_df = pd.DataFrame(demand_rows_HD)

  ################################################################
  ################################################################
  ################################################################
  # --- 4. 엑셀 파일 한 번에 저장하기 ---


  print(f"'{file_name}' 파일 생성을 시작합니다...")

  with pd.ExcelWriter(file_name, engine='openpyxl') as writer:
      # 시트 '17'에 s_rk 데이터 저장 (A2:C51)
      s_rk.to_excel(
          writer,
          sheet_name='17',
          startrow=1,
          startcol=0,
          header=False,
          index=False
      )
      print(" -> '17' 시트 저장 완료")

      # 시트 'plant'에 plant_candidates 데이터 저장 (A1:C3)
      plant_candidates.to_excel(
          writer,
          sheet_name='plant',
          startrow=0,
          startcol=0,
          header=False,
          index=False
      )
      print(" -> 'plant' 시트 저장 완료")

      # 시트 'customer'에 customer_location 데이터 저장 (A1:A50)
      customer_location.to_excel(
          writer,
          sheet_name='customer',
          startrow=0,
          startcol=0,
          header=False,
          index=False
      )
      print(" -> 'customer' 시트 저장 완료")

      # 시트 'dc'에 dc_candidates 데이터 저장 (A1:A10)
      dc_candidates.to_excel(
          writer,
          sheet_name='dc',
          startrow=0,
          startcol=0,
          header=False,
          index=False
      )
      print(" -> 'dc' 시트 저장 완료")

      # 시트 '6'에 mu_df 데이터 저장 (A2:C51)
      mu_df.to_excel(
          writer,
          sheet_name='6',
          startrow=1,
          startcol=0,
          header=False,
          index=False
      )
      print(" -> '6' 시트 저장 완료")

      # 시트 '5'에 mu_hat_df 데이터 저장 (A2:C51)
      mu_hat_df.to_excel(
          writer,
          sheet_name='5',
          startrow=1,
          startcol=0,
          header=False,
          index=False
      )
      print(" -> '5' 시트 저장 완료")

      # 시트 '2'에 flat_df 데이터 저장 (A1:D91)
      flat_df.to_excel(
          writer,
          sheet_name='2',
          startrow=0,
          startcol=0,
          header=True,   # 열 이름(k,i,j,D_kij) 포함
          index=False
      )
      print(" -> '2' 시트 저장 완료")

      # 시트 '3'에 D_jr 데이터 저장 (A1:AX11)
      D_jr.to_excel(
          writer,
          sheet_name='3',
          startrow=0,
          startcol=0,
          header=True,  # 열 이름(r1..r50) 포함
          index=False   # 행 이름(dc1..dc10) 제외
      )
      print(" -> '3' 시트 저장 완료")

      # 시트 '7'에 sheet_7_df 데이터 저장 (B2:D4)
      sheet_7_df.to_excel(
          writer,
          sheet_name='7',
          startrow=1,   # 엑셀 2행
          startcol=1,   # 엑셀 B열
          header=False,
          index=False
      )
      print(" -> '7' 시트 저장 완료")

      # 시트 '8'에 sheet_8_df 데이터 저장 (B1:B10)
      sheet_8_df.to_excel(
          writer,
          sheet_name='8',
          startrow=0,   # 엑셀 1행
          startcol=1,   # 엑셀 B열
          header=False,
          index=False
      )
      print(" -> '8' 시트 저장 완료")

      # 시트 '9'에 sheet_9_df 데이터 저장 (B2:D4)
      sheet_9_df.to_excel(
          writer,
          sheet_name='9',
          startrow=1,   # 엑셀 2행
          startcol=1,   # 엑셀 B열
          header=False,
          index=False
      )
      print(" -> '9' 시트 저장 완료")

      # 시트 '10'에 sheet_10_df 데이터 저장 (B1:B10)
      sheet_10_df.to_excel(
          writer,
          sheet_name='10',
          startrow=0,   # 엑셀 1행
          startcol=1,   # 엑셀 B열
          header=False,
          index=False
      )
      print(" -> '10' 시트 저장 완료")

      # 시트 '11'에 random_d_kij 데이터 저장 (D2:D91)
      sheet_11_df.to_excel(
          writer,
          sheet_name='11',
          startrow=1,   # 엑셀 2행
          startcol=3,   # 엑셀 D열
          header=False,
          index=False
      )
      print(" -> '11' 시트 저장 완료")

      # 시트 '12'에 sheet_12_df 데이터 저장 (A2:AX11)
      sheet_12_df.to_excel(
          writer,
          sheet_name='12',
          startrow=1,   # 엑셀 2행
          startcol=0,   # 엑셀 A열
          header=False,
          index=False
      )
      print(" -> '12' 시트 저장 완료")

      # 시트 '13'에 sheet_13_df 데이터 저장 (B2:B11)
      sheet_13_df.to_excel(
          writer,
          sheet_name='13',
          startrow=1,   # 엑셀 2행
          startcol=1,   # 엑셀 B열
          header=False,
          index=False
      )
      print(" -> '13' 시트 저장 완료")

      # 시트 '14'에 sheet_14_df 데이터 저장 (B2:B11)
      sheet_14_df.to_excel(
          writer,
          sheet_name='14',
          startrow=1,   # 엑셀 2행
          startcol=1,   # 엑셀 B열
          header=False,
          index=False
      )
      print(" -> '14' 시트 저장 완료")

      # 시트 '15'에 sheet_15_df 데이터 저장 (B2:D4)
      sheet_15_df.to_excel(
          writer,
          sheet_name='15',
          startrow=1,   # 엑셀 2행
          startcol=1,   # 엑셀 B열
          header=False,
          index=False
      )
      print(" -> '15' 시트 저장 완료")

      # 시트 'demand'에 sheet_demand_df 데이터 저장 (A1:C3)
      sheet_demand_HD_df.to_excel(
          writer,
          sheet_name='demand_HD',
          startrow=0,
          startcol=0,
          header=False,
          index=False
      )
      sheet_demand_LD_df.to_excel(
          writer,
          sheet_name='demand_LD',
          startrow=0,
          startcol=0,
          header=False,
          index=False
      )
      sheet_demand_MD_df.to_excel(
          writer,
          sheet_name='demand_MD',
          startrow=0,
          startcol=0,
          header=False,
          index=False
      )
      print(" -> 'demand' 시트 저장 완료")

  print(f"✅ '{file_name}' 파일에 모든 시트 저장을 완료했습니다.")
```

```python
import pandas as pd
import re
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# --- 1) 엑셀 파일 로드 및 시트별 DataFrame 생성 ---
excel_path = '/content/drive/MyDrive/First Paper/newdata/real_data/data_5.xlsx'
xls = pd.ExcelFile(excel_path)

plant_df    = pd.read_excel(xls, sheet_name='plant',    header=None)
customer_df = pd.read_excel(xls, sheet_name='customer', header=None)
dc_df       = pd.read_excel(xls, sheet_name='dc',       header=None)

# --- 2) 좌표 문자열 파싱 함수 정의 ---
# "(np.int64(425), np.int64(224))" → "(425, 224)"로 바꾼 뒤 숫자 추출
def parse_coord(s):
    s = str(s)
    # 1) 타입 래퍼 제거
    s = s.replace('np.int64(', '') \
         .replace('np.float64(', '')  # 필요하다면 여기에 다른 래퍼도 추가
    # 2) 남은 모든 ')' 제거
    s = s.replace(')', '')
    # 3) 숫자만 추출
    nums = re.findall(r'-?\d+', s)
    if len(nums) >= 2:
        return (int(nums[0]), int(nums[1]))
    return None

# --- 3) Plant 후보 파싱 (k1, k2, k3 그룹) ---
group_names = ['k1', 'k2', 'k3']
plant_candidates = {
    g: plant_df[col].dropna()
                      .map(parse_coord)
                      .dropna()
                      .tolist()
    for col, g in enumerate(group_names)
}

# --- 4) Customer 위치 파싱 ---
customer_points = customer_df[0].dropna() \
                                .map(parse_coord) \
                                .dropna() \
                                .tolist()

# --- 5) DC 후보 파싱 ---
dc_points = dc_df[0].dropna() \
                    .map(parse_coord) \
                    .dropna() \
                    .tolist()

# --- 6) 플로팅 시작 ---
fig, ax = plt.subplots(figsize=(10, 10))

# 도넛 모양 외부 영역 연한 파란색으로 칠하기 (zorder=0)
ax.add_patch(patches.Rectangle((0,   0),   1000, 200,  facecolor='lightblue', alpha=0.5, zorder=0))
ax.add_patch(patches.Rectangle((0,   800), 1000, 200,  facecolor='lightblue', alpha=0.5, zorder=0))
ax.add_patch(patches.Rectangle((0,   200), 200,  600,  facecolor='lightblue', alpha=0.5, zorder=0))
ax.add_patch(patches.Rectangle((800, 200), 200,  600,  facecolor='lightblue', alpha=0.5, zorder=0))

# Plant 그룹별 색상·마커 설정
color_map  = {'k1': 'red',   'k2': 'green', 'k3': 'blue'}
marker_map = {'k1': 'o',     'k2': 's',     'k3': '^'}

for group, pts in plant_candidates.items():
    xs = [p[0] for p in pts]
    ys = [p[1] for p in pts]
    ax.scatter(xs, ys,
               c=color_map[group],
               marker=marker_map[group],
               label=f'plant {group}',
               s=100)

# Customer 산점도 (검은 점)
cust_x = [p[0] for p in customer_points]
cust_y = [p[1] for p in customer_points]
ax.scatter(cust_x, cust_y,
           c='black',
           marker='.',
           alpha=0.7,
           label='customers')

# DC 후보 산점도 (주황색 ×)
dc_x = [p[0] for p in dc_points]
dc_y = [p[1] for p in dc_points]
ax.scatter(dc_x, dc_y,
           c='orange',
           marker='x',
           s=100,
           label='dc candidates')

# 축·제목·그리드·종횡비 설정
ax.set_xlim(0, 1000)
ax.set_ylim(0, 1000)
ax.set_xlabel('X coordinate')
ax.set_ylabel('Y coordinate')
ax.set_title('Plant Groups, Customer Locations, and DC Candidates')
ax.grid(True, linestyle='--', alpha=0.5)
ax.set_aspect('equal', adjustable='box')

# 범례를 그래프 밖에 배치하고 레이아웃 조정
ax.legend(loc='center left', bbox_to_anchor=(1.05, 0.5))
plt.tight_layout(rect=[0, 0, 0.85, 1])

plt.savefig('location_plot.png',
            dpi=300,
            bbox_inches='tight')

plt.show()
```

```python
import pandas as pd
import re
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.patches as patches

# === 1) 엑셀에서 좌표 읽기 & 파싱 함수 정의 ===
excel_path = '/content/drive/MyDrive/First Paper/newdata/real_data/data_5.xlsx'  # 실제 경로로 바꿔주세요
xls = pd.ExcelFile(excel_path)

plant_df    = pd.read_excel(xls, sheet_name='plant',    header=None)
customer_df = pd.read_excel(xls, sheet_name='customer', header=None)
dc_df       = pd.read_excel(xls, sheet_name='dc',       header=None)

def parse_coord(s):
    """
    "(np.int64(425), np.int64(224))" → "(425, 224)"로 바꾼 뒤 (int, int) 튜플 리턴
    """
    s = str(s).replace('np.int64(', '') \
               .replace('np.float64(', '') \
               .replace(')', '')
    nums = re.findall(r'-?\d+', s)
    return (int(nums[0]), int(nums[1])) if len(nums) >= 2 else None

# === 2) Plant 후보 파싱 (3×3 그룹 → 총 9개) ===
group_names = ['k1','k2','k3']
plant_candidates = {
    g: plant_df[col].dropna()
                     .map(parse_coord)
                     .dropna()
                     .tolist()
    for col, g in enumerate(group_names)
}
# 순서대로 평탄화하여 plant_points[0..8]에 매핑
plant_points = [pt for g in group_names for pt in plant_candidates[g]]

# === 3) Customer·DC 파싱 ===
customer_points = (
    customer_df[0]
    .dropna()
    .map(parse_coord)
    .dropna()
    .tolist()
)
dc_points = (
    dc_df[0]
    .dropna()
    .map(parse_coord)
    .dropna()
    .tolist()
)  # 10개

# === 4) Best Chromosome 멀티라인 파싱 ===
best_chr_lines = []
with open('/content/drive/MyDrive/First Paper/5_data_HD/5_HD_All.txt', 'r') as f:
    for line in f:
        if 'Best Chromosome' in line:
            best_chr_lines.append(line[line.find('['):].strip())
            if ']' in line:
                break
            for cont in f:
                best_chr_lines.append(cont.strip())
                if ']' in cont:
                    break
            break

best_chr_str = ' '.join(best_chr_lines)
nums = re.findall(r'-?\d+', best_chr_str)
best_chr = list(map(int, nums))

# === 5) 연결 정보 및 open 여부 판단 ===
conn = np.array(best_chr[:90]).reshape((9, 10))
open_plants = {i for i in range(9)  if conn[i].sum() > 0}
open_dcs    = {j for j in range(10) if conn[:, j].sum() > 0}

# === 6) 시각화 ===
fig, ax = plt.subplots(figsize=(10, 10))

# 6.1) “도넛” 외곽 영역
ax.add_patch(patches.Rectangle((0,   0),   1000, 200,  facecolor='lightblue', alpha=0.5, zorder=0))
ax.add_patch(patches.Rectangle((0,   800), 1000, 200,  facecolor='lightblue', alpha=0.5, zorder=0))
ax.add_patch(patches.Rectangle((0,   200), 200,  600,  facecolor='lightblue', alpha=0.5, zorder=0))
ax.add_patch(patches.Rectangle((800, 200), 200,  600,  facecolor='lightblue', alpha=0.5, zorder=0))

# 6.2) Customer 분포
cust_x = [p[0] for p in customer_points]
cust_y = [p[1] for p in customer_points]
ax.scatter(cust_x, cust_y, c='black', marker='.', alpha=0.7, label='customers')

# 6.3) Plant↔DC 연결선 (edge=1)
for i, (x0, y0) in enumerate(plant_points):
    for j, (x1, y1) in enumerate(dc_points):
        if conn[i, j] == 1:
            ax.annotate(
                "",                # 텍스트 없이
                xy=(x1, y1),       # 화살표 꼬리→머리(DC 위치)
                xytext=(x0, y0),   # 화살표 머리→꼬리(plant 위치)
                arrowprops=dict(
                    arrowstyle="->",   # 단방향 화살표
                    linewidth=2,
                    color="gray",
                    alpha=1.0
                )
            )

# 6.4) Plant 노드: 그룹별 색상·모양 + open/closed 투명도
color_map  = {'k1':'red',   'k2':'green', 'k3':'blue'}
marker_map = {'k1':'o',     'k2':'s',     'k3':'^'}
group_size = len(plant_candidates[group_names[0]])
plotted_groups = set()

for idx, (x, y) in enumerate(plant_points):
    grp = group_names[idx // group_size]
    alpha = 1.0 if idx in open_plants else 0.3
    label = f'plant {grp}' if grp not in plotted_groups else None
    if label:
        plotted_groups.add(grp)
    ax.scatter(x, y,
               c=color_map[grp],
               marker=marker_map[grp],
               edgecolor='black',
               s=120,
               alpha=alpha,
               label=label)

# 6.5) DC 노드: open/closed 표시
open_plotted = False
closed_plotted = False
for j, (x, y) in enumerate(dc_points):
    if j in open_dcs:
        if not open_plotted:
            ax.scatter(x, y, c='orange', marker='x', s=120, alpha=1.0, label='open DCs')
            open_plotted = True
        else:
            ax.scatter(x, y, c='orange', marker='x', s=120, alpha=1.0)
    else:
        if not closed_plotted:
            ax.scatter(x, y, c='orange', marker='x', s=120, alpha=0.3, label='closed DCs')
            closed_plotted = True
        else:
            ax.scatter(x, y, c='orange', marker='x', s=120, alpha=0.3)

# 6.6) 축·제목·그리드·비율·범례 설정
ax.set_xlim(0, 1000)
ax.set_ylim(0, 1000)
ax.set_xlabel('X coordinate')
ax.set_ylabel('Y coordinate')
ax.set_title('Plant–DC Connections (Best Chromosome)')
ax.grid(True, linestyle='--', alpha=0.5)
ax.set_aspect('equal', adjustable='box')
ax.legend(loc='center left', bbox_to_anchor=(1.05, 0.5))

plt.tight_layout(rect=[0, 0, 0.85, 1])
plt.show()
```

```python
import numpy as np, math

def phi(x):      # standard normal PDF
    return math.exp(-0.5*x*x) / math.sqrt(2*math.pi)

def Phi(x):      # standard normal CDF via erf
    return 0.5 * (1 + math.erf(x / math.sqrt(2)))

def truncated_mean(mu, sigma, a=1.0, b=3.0):
    alpha = (a - mu) / sigma
    beta  = (b - mu) / sigma
    Z = Phi(beta) - Phi(alpha)
    return mu + sigma * (phi(alpha) - phi(beta)) / Z

def find_mu_for_mean(target_mean=1.25, sigma=0.25, a=1.0, b=3.0):
    lo, hi = a - 3*sigma, target_mean     # mu must lie below target_mean
    for _ in range(100):
        mid = 0.5 * (lo + hi)
        if truncated_mean(mid, sigma, a, b) > target_mean:
            hi = mid
        else:
            lo = mid
    return 0.5 * (lo + hi)

sigma = 0.
mu    = find_mu_for_mean(sigma=sigma)
print(f"mu ≈ {mu:.3f},   sigma = {sigma},   mean = {truncated_mean(mu, sigma):.2f}")
```
